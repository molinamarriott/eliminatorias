# Introducción a Pandas

Pandas es una biblioteca poderosa y flexible para la manipulación y análisis de datos en Python. Ofrece estructuras de datos como DataFrames y Series para trabajar con datos tabulares de manera eficiente.

### 1. Instalación y Carga de Pandas

```python
# Si no tienes pandas instalado, puedes usar el siguiente comando:
# !pip install pandas

import pandas as pd
```

### 2. Creación de un DataFrame

2023-08-11-dataframes.jpg

```python
# Creando un DataFrame desde un diccionario
datos = {
    'Nombre': ['Ana', 'Luis', 'María', 'Pedro', 'Juana', 'Marta'],
    'Edad': [23, 30, 25, 40,60,70],
    'Ciudad': ['Madrid', 'Barcelona', 'Valencia', 'Sevilla', 'Guayaquil', 'Quito']
}

df = pd.DataFrame(datos)
df
```

### 3. Lectura de Archivos

```python
# Leer un archivo CSV (suponiendo que tienes un archivo 'archivo.csv')
# df = pd.read_csv('archivo.csv')
```

### 4. Exploración de Datos

```python
# Explorando el DataFrame
print(df.head())  # Muestra las primeras 5 filas
print(df.describe())  # Resumen estadístico de las columnas numéricas
print(df.info())  # Información del DataFrame
```

### 5. Selección de Datos

```python
# Seleccionando columnas
print(df['Nombre'])  # Selecciona la columna 'Nombre'

# Seleccionando filas por índice
print(df.iloc[1])  # Selecciona la segunda fila (índice 1)

# Selección condicional
print(df[df['Edad'] > 25])  # Filtra filas donde la edad es mayor que 25
```

### 6. Modificación de Datos

```python
# Agregar una nueva columna
df['Salario'] = [30000, 40000, 35000, 50000,10000, 800]
df
```

```python
# Modificar valores
df.loc[0, 'Edad'] = 24
df
```

### 7. Operaciones con Datos

```python
# Operaciones matemáticas
df['Edad'] = df['Edad'] + 1  # Incrementa en 1 la edad de todos
df['Salario_aumento'] = df['Salario'] * 1.1  # Aumenta el salario en un 10%
df
```

### 9. Eliminación de Filas o Columnas

```python
# Eliminar una columna
df.drop('Salario', axis=1, inplace=True)
df
```

```python
# Eliminar una fila
df.drop(1, axis=0, inplace=True)
df
```
