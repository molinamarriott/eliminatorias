# Introducción a los Conceptos Básicos de Python

### 1. Variables y Tipos de Datos

```python
# Ejemplo de variables y tipos de datos
nombre = "Ana"  # Tipo cadena (string)
edad = 25       # Tipo entero (int)
altura = 1.70   # Tipo flotante (float)
es_estudiante = True  # Tipo booleano

# Imprimiendo valores
print(f"Nombre: {nombre}, Edad: {edad}, Altura: {altura}, Estudiante: {es_estudiante}")
```

### 2. Operaciones Básicas

```python
# Ejemplo de operaciones básicas
suma = 5 + 3
resta = 5 - 3
producto = 5 * 3
division = 5 / 3
potencia = 5 ** 2

print(f"Suma: {suma}, Resta: {resta}, Producto: {producto}, División: {division}, Potencia: {potencia}")
```

### 3. Estructuras de Control

```python
# Condicionales (if, elif, else)
edad = 20
if edad >= 18:
    print("Eres mayor de edad.")
elif edad >= 13:
    print("Eres adolescente.")
else:
    print("Eres un niño.")
```

## Bucles (for, while)

```python
print("Bucle for:")
for i in ['a','b','c','d']:
    print(i)
```

```python
# Bucle while
print("\nBucle while:")
contador = 0
while contador < 3:
    print(contador)
    contador += 1
```

### 4. Funciones

```python
# Definiendo y llamando a una función
def saludar(nombre):
    return f"Hola, {nombre}!"

mensaje = saludar("Ana")
print(mensaje)
```

### 5. Listas y Sets

## Listas
1. Ordenada: esto quiere decir que los elementos dentro de ella están indexados y se accede a ellos a través de una locación indexada.
2. Editable: los elementos dentro de una lista pueden editarse, añadir nuevos o eliminar los que ya tiene.
3. Dinámica: las listas pueden contener diferentes tipos de datos y hasta de objetos. Esto significa que pueden soportar paquetes multidimensionales de datos, como un array o muchos objetos.
4. No única: esencialmente, esto quiere decir que la lista puede contener elementos duplicados sin que arroje un error.

```python
# Ejemplo de listas
frutas = ["manzana", "plátano", "naranja"]
print(frutas[0])  # Accediendo al primer elemento
frutas.append("fresa")  # Agregando un nuevo elemento
print(frutas)
```

## Sets



1. Los elementos de un set son único, lo que significa que no puede haber elementos duplicados
2. Los set son desordenados, lo que significa que no mantienen el orden de cuando son declarados.
3. Sus elementos deben ser inmutables.



```python
s = set([5, 4, 6, 8, 8, 1])
print(s)       #{1, 4, 5, 6, 8}
print(type(s)) #<class 'set'>
```

### 6. Diccionarios
Los diccionarios en Python son una estructura de datos que permite almacenar su contenido en forma de llave y valor.

1. Son dinámicos, pueden crecer o decrecer, se pueden añadir o eliminar elementos.
2. Son indexados, los elementos del diccionario son accesibles a través del key.
3. Son anidados, un diccionario puede contener a otro diccionario en su campo value.

```python
# Ejemplo de diccionario
persona = {"nombre": "Ana", "edad": 25}
print(persona["nombre"])  # Accediendo al valor de la clave 'nombre'
persona["edad"] = 26  # Modificando un valor
print(persona)
```
